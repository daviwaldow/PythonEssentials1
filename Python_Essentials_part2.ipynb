```python
#Exemplo 1 - programa para calcular mdia de altura de 5 pessoas em metros.
alturas = []
for i in range(1, 6):
    altura = float(input(f"Digite a altura da pessoa {i} em metros: "))
    alturas.append(altura)

    media_altura = sum(alturas) / len(alturas)
    print(f"\nA mdia de altura das 5 pessoas : {media_altura:.2f} metros")
```

```python
#Exemplo 2 - Programa para calcular raiz quadrada de um numero usando biblioteca.
import math
numero = float(input("Digite um nmero: "))
raiz_quadrada = math.sqrt(numero)
print(f"A raiz quadrada de {numero} : {raiz_quadrada}")
```

```python
#Programa para calcular x = pi.r
import math
raio = float(input("Digite o raio do crculo: "))
area = math.pi * raio ** 2
print(f"A rea do crculo : {area:.2f}")
```

```python
#Programa para calcular a mdia de aprovao de um estudante de universidade.

print("Digite as notas do estudante:")
nota1 = float(input("Nota 1: "))
nota2 = float(input("Nota 2: "))
nota3 = float(input("Nota 3: "))

media = (nota1 + nota2 + nota3) / 3

if media >= 7:
    print(f"O estudante foi aprovado com mdia {media:.2f}")
else:
    print(f"O estudante foi reprovado com mdia {media:.2f}")
```

```python
# Mini Jogo de Pedra, Papel, Tesoura. Utilizando bibliotecas de aleatoriedade, widgets e display para o Google Colabs
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

def jogar_pedra_papel_tesoura(escolha_jogador):
    opcoes = ['Pedra', 'Papel', 'Tesoura']
    computador = random.randint(0, 2)

    # Mostrar escolhas
    clear_output()
    print(f"Voc escolheu: {opcoes[escolha_jogador]}")
    print(f"O computador escolheu: {opcoes[computador]}")

    # Determinar resultado
    if escolha_jogador == computador:
        print("Empate!")
    elif (escolha_jogador == 0 and computador == 2) or \
         (escolha_jogador == 1 and computador == 0) or \
         (escolha_jogador == 2 and computador == 1):
        print("Parabns, voc ganhou!")
    else:
        print("Voc perdeu. Tente novamente!")

    # Boto para jogar novamente
    botao_reiniciar = widgets.Button(description="Jogar novamente")
    botao_reiniciar.on_click(lambda x: iniciar_jogo())
    display(botao_reiniciar)

def iniciar_jogo():
    clear_output()
    print("Bem-vindo ao jogo de Pedra, Papel e Tesoura!")
    print("Escolha uma das opes abaixo:")

    # Criar botes
    botao_pedra = widgets.Button(description="Pedra")
    botao_papel = widgets.Button(description="Papel")
    botao_tesoura = widgets.Button(description="Tesoura")

    # Conectar botes s funes
    botao_pedra.on_click(lambda x: jogar_pedra_papel_tesoura(0))
    botao_papel.on_click(lambda x: jogar_pedra_papel_tesoura(1))
    botao_tesoura.on_click(lambda x: jogar_pedra_papel_tesoura(2))

    # Mostrar botes
    display(widgets.HBox([botao_pedra, botao_papel, botao_tesoura]))

# Iniciar o jogo
iniciar_jogo()
```
